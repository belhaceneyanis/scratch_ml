```python
import numpy as np
import matplotlib.pyplot as plt
```

Linear Regression

```python
#y = ax+b
#multiplication of the matrix
# [x,1][a] = [y,1]
#      [b] 

#we start by generating the data 


x = np.arange(30)
y = 5*x + 23
#visuals
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^2')
plt.show()
```

```python

def linear_regression(x,y):
    #y = ax+b
    xx = np.mean(x)
    yy = np.mean(y)
    num = np.sum((x -xx)  *  (y -yy))
    denum = np.sum((x-xx)**2)
    a = num / denum
    b = yy - a*xx
    return a , b
```

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = 5*x + 23')
plt.show()


a, b = linear_regression(x,y)

c = np.arange(30)
d = a*c + b 

plt.scatter(c,d)
plt.xlabel('c')
plt.ylabel('d')
plt.title('c = a*c + b')
plt.show()
```
